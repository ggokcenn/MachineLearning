```python
import mglearn
import matplotlib.pyplot as plt
#Forge Dataset(synthetic)
X,y = mglearn.datasets.make_forge()
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.legend(["Class 0","Class 1"],loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
```

```python
import mglearn
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train,y_train)
```

```python
print("Test set predictions: {}".format(clf.predict(X_test)))
```

```python
print("Test set accuracy: {:.2f}".format(clf.score(X_test,y_test)))
```

```python
fig,axes= plt.subplots(1,3,figsize=(10,3))
for n_neighbors, ax in zip([1,3,9],axes):
    clf= KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf,X,fill=True,eps=0.5,ax=ax,alpha=.4)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc=3)
#zip kullanılarak oluşturulmuş bir for döngüsü
#verilen iki listeyi birleştirir ve döngüye çift girer
#mesela [1,3,9] ve axes(ax1,ax2,ax3)
#[1,ax1],[3,ax2],[9,ax3] şeklinde döngüye girerler.
```

Using single neighbor result in a decision boundary that follows the training data closely. Considering more and more neighbors leads a smoother decision boundary. A smoother boundary corresponds to a simpler model. 
Few neighbors=high model complexity.
Many neighbors= low model complexity
